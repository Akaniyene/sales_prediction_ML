# ML

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import make_scorer
from sklearn import tree
```

```python
df_features = new_data[["Store","Dept","Type_A","Type_B","Type_C", "Size", "IsHoliday","Day", "Week","Month", "Year"]]
df_label = new_data["Weekly_Sales"]
```

```python
df_features.info()
df_label.shape
```

```python
# Split the data into training and test set (70% and 30%) respestivcy

x_train, x_test, y_train, y_test = train_test_split(df_features, df_label, test_size = 0.3, random_state = 42)

print('Size of training set = ', x_train.shape)
print('Size of training label = ', y_train.shape)

print('\nSize of test set = ', x_test.shape)
print('Size of test label = ', y_test.shape)
```

```python
rf = RandomForestRegressor(random_state = 42)
```

We want to see what are the parameters required

```python
rf.get_params()
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 200, num = 5)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(100, 200, num = 5)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 4, 8, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
rf = RandomForestRegressor(n_jobs = -1, random_state = 42)
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 10, scoring= 'neg_mean_absolute_error', cv = 3, verbose = 10, random_state=42, n_jobs = -1)
#Fit the random search model
rf_random.fit(x_train, y_train)
```

```python
rf_random.best_params_
```

```python
base_model = RandomForestRegressor(n_jobs=-1, random_state = 42)
base_model.fit(x_train, y_train)
```

```python
predictions = base_model.predict(x_test)
```

```python
mea_base = mean_absolute_error(y_test, predictions)
print(mea_base)
```

```python
best_random = rf_random.best_estimator_
```

```python
predictions_best = best_random.predict(x_test)
```

```python
mea_best = mean_absolute_error(y_test , predictions_best)
print(mea_best)
```
